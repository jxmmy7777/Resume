```python
%matplotlib inline
import matplotlib.pyplot as plt
inputfile=open('data.csv',"r")
data=inputfile.read().split()
x=[]
y=[]
data2=[]
for i in data:
    string=(i.split(","))
    x.append(float(string[0]))
    y.append(float(string[1]))
fig=plt.figure()
plt.plot(x,y,'ro')
plt.axis([-30,30,-30,30])
plt.show()
fig.savefig('points_1.pdf')
#2
averagex=0
averagey=0
sigmax=0
sigmay=0
c=0
for a in range(len(x)):
    averagex=averagex+x[a]/len(x)
    averagey=averagey+y[a]/len(y)
for a in range(len(x)):
    sigmax=sigmax+(x[a]-averagex)**2
    sigmay=sigmay+(y[a]-averagey)**2
    c=c+(x[a]-averagex)*(y[a]-averagey)

r=c/((sigmax**0.5)*(sigmay**0.5))
print r
sigmax=(sigmax/len(x))**0.5
sigmay=(sigmay/len(y))**0.5

x2=x
y2=[averagey+r*sigmay/sigmax*(num-averagex) for num in x2]
fig=plt.figure()
plt.plot(x,y,'ro')
plt.plot(x2,y2)
plt.axis([-30,30,-30,30])
plt.show()
fig.savefig('line_1.pdf')
#3
Z_x=[]
Z_y=[]
for a in x:
    c=(a-averagex)/sigmax
    Z_x.append(c)
for d in y:
    i=(d-averagey)/sigmay
    Z_y.append(i)
print sigmax
print sigmay
print r
#4
fig=plt.figure()
plt.plot(Z_x,Z_y,'ro')
plt.axis([-3,3,-3,3])
plt.show()
fig.savefig('points_2.pdf')
#5

x5=Z_x
y5=[r*num for num in x5]
fig=plt.figure()
plt.plot(x5,y5)
plt.plot(Z_x,Z_y,'ro')
plt.show()
fig.savefig('line_2.pdf')
print r
```
